```python
from scipy.io import wavfile
import scipy.signal as sps
wav_fname = "C:\\Users\\vinay\\Downloads\\FEIS_v1_1\\augmented_wavs\\01\\wavs\\f_0.wav"
samplerate, data = wavfile.read(wav_fname)
new_number_of_samples = 8000
new_data = sps.resample(data, new_number_of_samples)
```

```python
print(samplerate, data, type(data), data.shape)
```

```python
print(new_data, new_data, type(new_data), new_data.shape)
```

```python
plt.figure(1)
plt.title("Original Signal Wave")
plt.plot(data)

plt.figure(2)
plt.title("Downsampled Signal Wave")
plt.plot(new_data)

plt.show()
```

```python
import IPython.display as ipd
ipd.display(ipd.Audio(data=data, rate=44100))
ipd.display(ipd.Audio(data=new_data, rate=8000))
```

```python
from scipy import signal
f, t, Zxx = signal.stft(data, samplerate, nperseg=256)
plt.pcolormesh(t, f, np.abs(Zxx), shading='gouraud')
plt.title('STFT Magnitude')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

```python
f, t, Zxx = signal.stft(new_data, 8000, nperseg=256)
plt.pcolormesh(t, f, np.abs(Zxx), shading='gouraud')
plt.title('STFT Magnitude')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

```python
f, t, Zxx = signal.stft(data, samplerate, window='boxcar', nperseg=512, nfft=512, noverlap=103)
plt.pcolormesh(t, f, np.abs(Zxx), shading='gouraud')
plt.title('STFT Magnitude')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```
